Now let us clean up the data.

```python
def FindNullCount (df, col):
    if col in df.columns:
        print "There are {0} missing items for column name {1}".format (df[col].isnull().sum(), col)
    else:
        print "Column {0} does not exist in the dataframe".format (col)

FindNullCount(df, "PassengerId")
FindNullCount(df, "Survived")
FindNullCount(df, "Pclass")
FindNullCount(df, "Name")
FindNullCount(df, "Sex")
FindNullCount(df, "Age")
FindNullCount(df, "SibSp")
FindNullCount(df, "Parch")
FindNullCount(df, "Ticket")
FindNullCount(df, "Fare")
FindNullCount(df, "Cabin")
FindNullCount(df, "Embarked")
    
```

```python
print df['Age'].mean()
print df['Age'].median()
print df['Age'].mode()
```

```python
# Now let us fix the missing items for Age. I prefer to choose the median as the missing at random. Morever there seems not much variation in mean, median and mode..
df['Age'].fillna(df['Age'].median(), inplace=True)

# Now let us fix the missing items with missing string" MISSING" for Cabin which is huge in number
df['Cabin'].fillna("MISSING", inplace=True)

# Now let us fix the missing items for Embarked wih values coming in S , C , Q
import random
df['Embarked'].fillna(random.choice(["S","C","Q"]), inplace=True)
```

```python
df_Dead =  df[df['Survived']==0]
df_Dead_Class1 = df_Dead[df_Dead['Pclass']==1]['Sex'].value_counts()
df_Dead_Class2 = df_Dead[df_Dead['Pclass']==2]['Sex'].value_counts()
df_Dead_Class3 = df_Dead[df_Dead['Pclass']==3]['Sex'].value_counts()

df_1 = pd.DataFrame([df_Dead_Class1,df_Dead_Class2,df_Dead_Class3])
df_1.index = ['Class1','Class2','Class3']
df_1.plot(kind='bar',stacked=True, figsize=(10,6))
plt.title('Distribution of people who died v/s the class')
plt.xlabel('Class in Titanic ')
plt.ylabel('People who Died')
plt.show()
```

<b>Now let us see if the location of the people who died played a role?</b>

```python
%matplotlib inline
dicData = {}
df_group_Embarked = df.groupby('Embarked')
for key, items in df_group_Embarked:
    group = df_group_Embarked.get_group(key)
    dead = group[group['Survived'] == 0]
    survived = group[group['Survived'] == 1]
    dicData[key] = [len(dead), len(survived)]
df_3 = pd.DataFrame.from_dict(dicData)
df_3 = df_3.T # transpose
df_3 = df_3.rename(columns={0: 'Dead', 1: 'Survived'},index={'C': 'Cherbourg','Q': 'Queenstown','S': 'Southampton'})
df_3.plot(kind='bar',stacked=True, figsize=(10,4), title = 'Distribution of people who died v/s where they embarked from')
plt.xlabel('Embarked From')
plt.ylabel('People who Died')
```

This graph shows that to some extend people who emabarked from Southampton died more compared to Queenstown or Cherbourg<br/>
So does this mean that people from Southampton were poorer? 
I caanot make any conclusion for this as the data provided as no means to show that...
